## Exercici 10

````python
# La teva resposta aquí

palabra = "bingo"
userpalabra=""
print("Adivina la palabra secreta de 5 letras. Tienes 10 intentos y el juego finaliza cuando pierdes o cuando adivines la palabra")
print(" ")
vidas=10
print("Tienes", vidas, "intentos.")
while vidas > 0:
    vidas=vidas-1
    for letra in palabra:
        if letra in userpalabra:
            print(letra,end=" ")
        else:
            print("_ ",end="")
    print(" ")
    print(" ")
    userletra=input("Introduce una letra: ")
    userpalabra+=userletra
    if userletra not in palabra:
        print("Letra incorrecta")
        print("Te quedan",+vidas," intentos")
    if vidas== 0:
        print("perdiste!")
    if palabra==userpalabra:
        print("b i n g o")
        break
print(" ")
print("Muchas gracias por jugar")
````

## Exercici 14

Modifica el programa anterior per a que la nota final s’obtingui de forma aleatòria tenin en compte la següent comanda rand = random.uniform(0,1). Aquesta comanda obté un nombre aleatori amb decimals entre 0 i 1. La nota del primer examen es multiplicarà per aquest factor i la d’aquest examen per la resta. S’ha de mostrar la nota final.

````python
# La teva resposta aquí

import random
nota1=float(input("¿Qué nota sacaste en el primer exámen?: "))
nota2=float(input("¿Qué nota crees que sacarás en el próximo exámen?: "))
rand = random.uniform(0,1)
print(rand)
print("La media aleatoria de tus notas es: ",((nota1*rand)+(nota2*(1-rand))))
````

## Exercici 15

Donada la següent llista = [39,62,24,71,40,53,12,99,20,111], desenvolupa un programa que generi una llista amb els nombres divisibles entre 3 i una altra llista amb els que no ho son. Un cop estan distribuits, la llista original s’ha de substituir per un 0 en tots els seus elements. Un cop acabat, es mostren totes les llistes.

````python
# La teva resposta aquí

llista = [39, 62, 24, 71, 40, 53, 12, 99, 20, 111]
print("La lista original es: ",llista)
div= []
nodiv= []
for x in llista:
    if x%3==0:
        div.append(x)
    else:
        nodiv.append(x)
for index in range (0,len(llista)):
    llista[index]=0
print("La lista de divisibles entre tres es: ",div)
print("La lista de no divisibles entre tres es: ",nodiv)
print("La lista original con ceros es: ",llista)
````
